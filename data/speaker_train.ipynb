```python
import pandas as pd
import numpy as np
import os
import re
```

```python
file = open('uet_train.txt')
```

```python
sentence  = file.readlines()
```

```python
sentence
```

```python
len(sentence)
```

```python
sentences=[]
dialogue_num=[]
speakers=[]
```

```python
flag=0
cnt=1

for i,s in enumerate(sentence):
    if s.startswith('"화자'):
        if flag==1:
            speakers.append(s)
        else:
            speakers.append(s)
            cnt=cnt+1
            flag=1

    else:
        sentences.append(s)
        dialogue_num.append(cnt)
        flag=0
```

```python
uet_train = pd.DataFrame()
```

```python
uet_train['sentence'] = sentences
```

```python
uet_train['d_num'] = dialogue_num
```

```python
uet_train['speaker'] = speakers
```

```python
uet_train['label'] = speakers
```

```python
uet_train['sentence'] = uet_train['sentence'].apply(lambda x: re.sub('"|\n','',x))
```

```python
uet_train['sentence'] = uet_train['sentence'].apply(lambda x: re.sub('([^,]),$','\\1',x)) # 마지막에 반점 삭제
```

```python
uet_train['speaker'] = uet_train['speaker'].apply(lambda x:x[1:4])
```

```python
uet_train['label'] = uet_train['label'].apply(lambda x: re.sub('"|\n','',x))
```

```python
uet_train['label'] = uet_train['label'].apply(lambda x:x[4:-1])
```
