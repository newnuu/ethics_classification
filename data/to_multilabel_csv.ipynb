# 데이터 개수 맞추기

```python
import pandas as pd
```

```python
data = pd.read_csv('../unethical_01.csv')
```

```python
data
```

```python
data[['범죄','혐오','선정','비도덕 아님','폭력','비난','욕설','차별']].sum()/9973
```

```python
data_crime = data[['sentence','범죄']]
```

```python
is_crime = data_crime['범죄']==1
```

```python
dc=data_crime[is_crime]
```

```python
data_crime = data_crime.append([dc]*((len(data_crime[data_crime['범죄']==0])//len(data_crime[data_crime['범죄']==1]))-1),ignore_index=True)
                  
                  
```

```python
data_crime.to_csv('../data_crime.csv')
```

```python
llist = ['범죄','혐오','선정','비도덕 아님','폭력','욕설','차별']
ll  = ['crime', 'hate', 'sexual', 'none', 'violent', 'curse','discrimination']

for i,lbl in enumerate(llist):
    data_label = data[['sentence',lbl]]
    is_true = data_label[lbl]==1
    temp=data_label[is_true]
    data_label = data_label.append([temp]*((len(data_label[data_label[lbl]==0])//len(data_label[data_label[lbl]==1]))),ignore_index=True)
    print(i,len(data_label[data_label[lbl]==0]),len(data_label[data_label[lbl]==1]))
    data_label.to_csv('../data_'+ll[i]+'.csv',index=False) 
    
data_blame = data[['sentence','비난']]
print(len(data_blame[data_blame['비난']==0]),len(data_blame[data_blame['비난']==1]))
data_blame.to_csv('../data_blame.csv',index=False)

```
