```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./train_ml2.csv')
```

```python
train = train[['sentence','d_num','speaker','label']]
```

```python
train
```

```python
sentence = train.groupby('d_num')['sentence'].apply(lambda x: ' '.join(x))
```

```python
label = train.groupby('d_num')['label'].apply(lambda x: ','.join(x))
```

```python
sentence
```

```python
label
```

```python
dialogue_merge = pd.merge(sentence,label,how='inner',on='d_num')
```

```python
dialogue_merge
```

```python
lbl = dialogue_merge['label'].str.split(',')
```

```python
lbl = lbl.apply(lambda x: list(set(x)))
```

```python
lbl
```

```python
lbl = pd.DataFrame(lbl)
```

```python
lbl
```

```python
def none(x):
    if len(x)>1 and '비도덕 아님' in x:
        x.remove('비도덕 아님')
    return x
```

```python
lbl = lbl['label'].apply(none)
```

```python
lbl
```

```python
lbl[lbl==['비도덕 아님']]
```

```python
lbl = lbl.apply(lambda x: ','.join(x))
```

```python
lbl = pd.DataFrame(lbl)
```

```python
lbl
```

```python
lblist = ['범죄', '선정', '비도덕 아님', '폭력', '비난', '혐오', '욕설', '차별']
```

```python
for lb in lblist:
    dialogue_merge[lb] = lbl['label'].str.contains(lb)
```

```python
dialogue_merge
```

```python
dialogue_merge = dialogue_merge.astype({'범죄':int,'혐오':int,'선정':int,'비도덕 아님':int,'폭력':int,'비난':int,'욕설':int,'차별':int})
```

```python
dialogue_merge = dialogue_merge[['sentence','범죄','혐오','선정','비도덕 아님','폭력','비난','욕설','차별']]
```

```python
dialogue_merge 
```

```python
dialogue_merge['sentence'][1]
```

```python
dialogue_merge['sentence'].apply(lambda x:len(x)).describe()
```

```python
dialogue_merge.to_csv('./dialogue_merge.csv',index=False)
```
