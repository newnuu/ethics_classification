```python
import pandas as pd
import numpy as np
import os
import re
```

```python
file = open('test2.txt')
```

```python
sentence  = file.readlines()
```

```python
len(sentence)
```

```python
sentences=[]
dialogue_num=[]
speakers=[]
```

```python
flag=0
cnt=1

for i,s in enumerate(sentence):
    if s.startswith('"화자'):
        if flag==1:
            speakers.append(s)
        else:
            speakers.append(s)
            cnt=cnt+1
            flag=1

    else:
        sentences.append(s)
        dialogue_num.append(cnt)
        flag=0
```

```python
uet_test = pd.DataFrame()
```

```python
uet_test['sentence'] = sentences
```

```python
uet_test['d_num'] = dialogue_num
```

```python
uet_test['speaker'] = speakers
```

```python
uet_test['label'] = speakers
```

```python
uet_test['sentence'] = uet_test['sentence'].apply(lambda x:x[1:-3])
```

```python
uet_test['speaker'] = uet_test['speaker'].apply(lambda x:x[1:4])
```

```python
uet_test['label'] = uet_test['label'].apply(lambda x: re.sub('"|\n','',x))
```

```python
uet_test['label'] = uet_test['label'].apply(lambda x:x[4:-1])
```
